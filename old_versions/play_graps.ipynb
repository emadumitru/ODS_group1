```python
data_graph = 'data/RI/'
data_map = 'data/RI/RI_tracts.shp'
name = 'Rhode Island'
```

```python
def run_graps(data_graph, data_map, name):
    G, full = load_graph(data_graph)
    # plot_graph(G, name)
    partitions = comprehensive_partitioning(G, 2)
    if partitions[1] is not None:
        plot_partitioned_graph(partitions[2], partitions[0], name + ' biggest subgraph')
        plot_districts_from_tracts(partitions[2], partitions[0], data_map, name + ' biggest subgraph')
        plot_partitioned_graph(G, partitions[1], name + ' multipe subgraphs')
        plot_districts_from_tracts(G, partitions[1], data_map, name + ' multipe subgraphs')
    else:
        plot_partitioned_graph(G, partitions[0], name)
        plot_districts_from_tracts(G, partitions[0], data_map, name)
```

```python
run_graps(data_graph, data_map, name)
```
